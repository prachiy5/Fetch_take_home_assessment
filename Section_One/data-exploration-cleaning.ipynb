
# **TRANSACTIONS DATASET CLEANING**

## **Transaction Dataset Cleaning Summary**

Steps taken to clean and standardize the `transactions` dataset before analysis.

1. **Converting Date Columns to Datetime**  
   - Converted `PURCHASE_DATE` and `SCAN_DATE` to `datetime` format for accurate time-based analysis.

2. **Standardizing the BARCODE Column**  
   - Converted barcode values stored as floats into string format while handling missing values.  
   - Ensured empty string values were replaced with `pd.NA` to maintain consistency in missing data handling.

3. **Cleaning the FINAL_QUANTITY Column**  
   - Replaced `"zero"` string values with `0.00`.  
   - Converted `FINAL_QUANTITY` to a numeric type, ensuring all values are correctly formatted.  
   - Used `errors="coerce"` to handle non-numeric values by converting them to NaN instead of raising errors.

4. **Creating Transaction Flags**  
   - Flagged transactions where quantity is `0` but sales value is positive.  
   - Identified transactions where quantity values are extremely high.

5. **Standardizing String Columns**  
   - Identified object (string) columns and applied standardization:  
     - Trimmed spaces.  
     - Converted values to lowercase for consistency.  
   - Replaced placeholder values (`"unknown"`, `"placeholder"`, `"null"`, `""`) with `pd.NA` for better missing value handling.

---

- **Note:** These steps ensure consistency and accuracy in transaction data before further analysis.  
- The following cells contain the implementation of these cleaning steps.


```python
# Displaying the first few rows
transactions.head()
```

```python
# Checking the data types of all columns
transactions.dtypes
```

```python
# Convert to datetime
transactions["PURCHASE_DATE"] = pd.to_datetime(transactions["PURCHASE_DATE"], errors="coerce")
print(transactions[["PURCHASE_DATE"]].head())
```

```python
# Convert to datetime
transactions["SCAN_DATE"] = pd.to_datetime(transactions["SCAN_DATE"], errors="coerce")
print(transactions[["PURCHASE_DATE"]].head())
```

```python
# Converting the BARCODE column to a string while handling missing values:
# - `float(x)`: Ensures that barcodes stored as float (e.g., 4.8001353664e+10) are converted properly.
# - `int(float(x))`: Removes decimal points from barcodes stored as floats.
# - `str(...)`: Converts the barcode to a string to maintain consistency.
# - `pd.notna(x)`: Ensures missing values (NaN) are handled and replaced with `pd.NA` instead of an error.
transactions["BARCODE"] = transactions["BARCODE"].apply(lambda x: str(int(float(x))) if pd.notna(x) else pd.NA)
transactions.head()
```

```python
# Replacing empty string values in the BARCODE column with pd.NA (missing value indicator).
# - Ensures that empty values are treated as missing data rather than empty strings.
# - Helps maintain consistency in handling missing barcodes for analysis and joins.

transactions["BARCODE"] = transactions["BARCODE"].replace("", pd.NA)
```

```python
## Counting the number of missing values (NaN or pd.NA) in the BARCODE column.
transactions['BARCODE'].isnull().sum()
```

```python
# Replacing the string value "zero" with 0.00 in the FINAL_QUANTITY column.

transactions["FINAL_QUANTITY"] = transactions["FINAL_QUANTITY"].replace("zero", 0.00)
```

```python
# Converting the FINAL_QUANTITY column to a numeric data type.
# - `pd.to_numeric(...)` ensures all values are properly converted to numbers.
# - `errors="coerce"` replaces non-numeric values with NaN instead of raising an error.
transactions["FINAL_QUANTITY"] = pd.to_numeric(transactions["FINAL_QUANTITY"], errors="coerce")
```

```python
#check data type
transactions.dtypes
```

```python
#check duplicate records
num_exact_duplicates = transactions.duplicated().sum()
print(f"Total number of exact duplicate records: {num_exact_duplicates}")
```

```python
#check total records after removing duplicates
transactions = transactions.drop_duplicates(keep="first")
print(f"Total records after removing exact duplicates: {transactions.shape[0]}")
```

```python
print(f"Remaining duplicate records: {transactions.duplicated().sum()}")
```

```python
#checking if total_quantity can have decimal values

# Count how many values are whole numbers vs decimals
whole_numbers = transactions[transactions["FINAL_QUANTITY"] % 1 == 0].shape[0]
decimals = transactions[transactions["FINAL_QUANTITY"] % 1 != 0].shape[0]

print(f"\nTotal Whole Number Quantities: {whole_numbers}")
print(f"Total Decimal Quantities: {decimals}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot to see correlation between FINAL_QUANTITY and FINAL_SALE
plt.figure(figsize=(8,5))
sns.scatterplot(x=transactions["FINAL_QUANTITY"], y=transactions["FINAL_SALE"])
plt.xlabel("FINAL_QUANTITY")
plt.ylabel("FINAL_SALE")
plt.title("Relationship between Quantity and Sale Price")
plt.show()
```

```python
# Find transactions where FINAL_QUANTITY is 0 but FINAL_SALE is not
zero_qty_sales = transactions[(transactions["FINAL_QUANTITY"] == 0.0) & (transactions["FINAL_SALE"] > 0)]
print(zero_qty_sales.head())
```

```python
# Check transactions with unusually high quantities
high_qty_sales = transactions[transactions["FINAL_QUANTITY"] > 100]
print(high_qty_sales.head())
```

```python
# Create a flag for transactions where quantity is 0 but sale is positive
transactions["ANOMALY_ZERO_QTY"] = (transactions["FINAL_QUANTITY"] == 0) & (transactions["FINAL_SALE"] > 0)
transactions.head()
```

```python
# Flag transactions where quantity is extremely high
transactions["ANOMALY_HIGH_QTY"] = transactions["FINAL_QUANTITY"] > 100
transactions.head()
```

```python
# Identify object (string) columns
string_columns = transactions.select_dtypes(include="object").columns


# Standardizing all string columns (strip spaces & lowercase)
for col in string_columns:
    transactions[col] = transactions[col].astype(str).str.strip().str.lower()

# Replace placeholders with NaN
transactions.replace(["unknown", "placeholder", "null", ""], pd.NA, inplace=True)

print(transactions[string_columns].sample(5))  # Random sample check

```

```python
# Assign a unique sequential Transaction_ID to each row in the 'transactions' DataFrame
# This ensures every transaction has a distinct identifier
transactions["TRANSACTION_ID"] = range(1, len(transactions) + 1)

# Check for any duplicate Transaction_IDs to verify uniqueness
print("Duplicate Transaction_IDs:", transactions["TRANSACTION_ID"].duplicated().sum())
```

```python
# Save the 'transactions' DataFrame to a CSV file named 'transactions_data_new.csv'
# The 'index=False' parameter ensures that the DataFrame index is not included in the CSV file
transactions.to_csv("transactions_data_new.csv", index=False)
```



# **PRODUCTS DATASET CLEANING**

## **Products Dataset Cleaning Summary**

Steps taken to clean and standardize the `products` dataset before analysis.

1. **Standardizing the BARCODE Column**  
   - Converted barcodes stored as floats to string format while ensuring missing values were handled properly.  

2. **Cleaning and Standardizing Category & Brand Columns**  
   - Standardized categorical columns: `"CATEGORY_1"`, `"CATEGORY_2"`, `"CATEGORY_3"`, `"CATEGORY_4"`, `"MANUFACTURER"`, and `"BRAND"`.  
   - Applied transformations:
     - Trimmed extra spaces.
     - Converted values to title case for better readability.
   - Replaced placeholder values:
     - `"Placeholder Manufacturer"` → `pd.NA`
     - `"Brand Not Known"` → `pd.NA`

3. **Removing Duplicates**  
   - Identified and removed exact duplicate records, keeping only the first occurrence.  
   - Ensured dataset uniqueness after duplicate removal.

4. **Generating Unique Product Identifiers**  
   - Created a new unique `PRODUCT_ID` column by assigning a sequential range of numbers.  
   - Verified that all product IDs were unique.

5. **Saving the Cleaned Dataset**  
   - Exported the cleaned `products` dataset to a CSV file (`products_data_new.csv`).  
   - Used `index=False` to ensure the DataFrame index was not included in the file.

---

- **Note:** These cleaning steps ensure that product data is properly formatted and standardized before further analysis.  
- The following cells contain the implementation of these cleaning steps.


```python
# Display the first five rows of the 'products' DataFrame to inspect its structure and initial data
products.head()
```

```python
# Display the data types of each column in the 'products' DataFrame
# This helps in understanding the structure of the dataset and identifying any necessary type conversions
products.dtypes
```

```python
# Convert the 'BARCODE' column to a string format while handling potential float values
# - Convert non-null values to float first, then to integer, and finally to string to remove decimal points
# - Preserve missing values as NaN (or pd.NA for better compatibility with newer pandas versions)
products["BARCODE"] = products["BARCODE"].apply(lambda x: str(int(float(x))) if pd.notna(x) else pd.NA)

# Display the first five rows to verify the transformation
products.head()

# Count and display the number of missing values in the 'BARCODE' column after conversion
products['BARCODE'].isnull().sum()
```

```python
# List of object columns to standardize
columns_to_clean = ["CATEGORY_1", "CATEGORY_2", "CATEGORY_3", "CATEGORY_4", "MANUFACTURER", "BRAND"]

# Function to clean text data
def clean_text(value):
    if pd.isna(value):  # Keep NaN values as pd.NA
        return pd.NA
    value = str(value).strip()  # Remove extra spaces
    value = value.title()  # Convert to title case (better readability)
    return value

# Apply cleaning function to all relevant columns
for col in columns_to_clean:
    products[col] = products[col].apply(clean_text)

# Verify the changes
print(products[columns_to_clean].head(10))  # Check cleaned data
```

```python
# Replace placeholder values with NaN
products["MANUFACTURER"].replace("Placeholder Manufacturer", pd.NA, inplace=True)
products["BRAND"].replace("Brand Not Known", pd.NA, inplace=True)

# Check if it worked
print(products[["MANUFACTURER", "BRAND"]].isna().sum())
```

```python
#check total null values in each column
products.isnull().sum()
```

```python
#check data type
products.dtypes
```

```python
# Find exact duplicates across all columns
exact_duplicates = products[products.duplicated(keep=False)]

# Display the duplicates sorted for better visibility
display(exact_duplicates.sort_values(by=products.columns.tolist()).head(20))
```

```python
# Count exact duplicate records
num_exact_duplicates = exact_duplicates.shape[0]

print(f"Total number of exact duplicate records: {num_exact_duplicates}")
```

```python
# Drop exact duplicates (keep first occurrence)
products = products.drop_duplicates(keep="first")

print(f"Total records after removing exact duplicates: {products.shape[0]}")
```

```python
# Check if any exact duplicates still exist
remaining_duplicates = products.duplicated().sum()
print(f"Remaining exact duplicate records: {remaining_duplicates}")
```

```python


# Create a heatmap to visualize missing values in the category columns of the 'products' DataFrame
# - 'CATEGORY_1', 'CATEGORY_2', 'CATEGORY_3', and 'CATEGORY_4' are checked for missing values
# - 'cmap="viridis"' sets the color map for better visibility
# - 'cbar=False' removes the color bar for simplicity
plt.figure(figsize=(10, 6))
sns.heatmap(products[["CATEGORY_1", "CATEGORY_2", "CATEGORY_3", "CATEGORY_4"]].isnull(), cmap="viridis", cbar=False)

# Set the title for the heatmap
plt.title("Missing Data Heatmap for Categories")

# Display the heatmap
plt.show()
```

```python
# Assign a unique sequential Product_ID to each row in the 'products' DataFrame
# This ensures every product has a distinct identifier
products["PRODUCT_ID"] = range(1, len(products) + 1)

# Check for any duplicate Product_IDs to verify uniqueness
print("Duplicate Product_IDs:", products["PRODUCT_ID"].duplicated().sum())
```

```python
# Save the 'products' DataFrame to a CSV file named 'products_data_new.csv'
# The 'index=False' parameter ensures that the DataFrame index is not included in the CSV file
products.to_csv("products_data_new.csv", index=False)
```

# **VISUALIZATIONS**

## **Summary of Visualizations Created**

1. **Top 10 Stores by Transaction Count (Cleaned Data)**
2. **Top 10 Product Categories by Sales**
---

These visualizations help in understanding transaction patterns across stores and product categories. The following cells contain the implementation of these visualizations.


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Merge datasets
merged_data = transactions.merge(users, left_on='USER_ID', right_on='ID', how='left').merge(products, on='BARCODE', how='left')
```

```python
# Check for missing values in STORE_NAME
missing_store_names = merged_data['STORE_NAME'].isna().sum()
print(f"Missing STORE_NAME values: {missing_store_names}")

# Drop missing store names before plotting (if needed)
filtered_data = merged_data.dropna(subset=['STORE_NAME'])

# Replot the visualization with cleaned data
plt.figure(figsize=(12, 6))
store_counts = filtered_data['STORE_NAME'].value_counts().head(10)
sns.barplot(x=store_counts.values, y=store_counts.index, palette="coolwarm")
plt.title('Top 10 Stores by Transaction Count (Cleaned)')
plt.xlabel('Number of Transactions')
plt.ylabel('Store Name')
plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Top 10 Product Categories by Sales
plt.figure(figsize=(12, 6))
top_categories = merged_data['CATEGORY_1'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette="magma")
plt.title('Top 10 Product Categories by Purchase Count')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```
